```python
# Importing Libraries

import pandas as pd
import numpy as np
```

```python
# Load data

energy = pd.read_csv('/content/drive/MyDrive/Data/energydata.csv')
```

```python
energy.head()
```

```python
# Statistical summary of the dataset

energy.describe()
```

```python
# Data description

energy.info()
```

```python
# Checking if there are any null values

energy.isna().sum()
```

```python
# Dropping date and lights variables

energy.drop(['date','lights'], axis=1, inplace=True)
```

```python
# Normalizing the dataset using MinMaxScaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalized_df = pd.DataFrame(scaler.fit_transform(energy), columns=energy.columns)
features_df = normalized_df.drop(columns=['Appliances'])
Appliances = normalized_df['Appliances']
```

```python
# Splitting data into train and test set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features_df, Appliances,
test_size=0.3, random_state=42)
```

```python
# Model

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

linear_model.fit(x_train, y_train) # Fitting the model

predicted_values = linear_model.predict(x_test) # Evaluating the model using the test set
```

```python
# Mean Squared Error

from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
# Residual of Sum Squares

import numpy as np

rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

```python
# Root Mean Squared Error

from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3) 
```

```python
# Coefficient of Determination

from sklearn.metrics import r2_score

r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2) 
```

```python
# This function returns the weight of every feature

def get_weights_df(linear_model, feat, col_name):
  weights = pd.Series(linear_model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['features_df', col_name]
  weights_df[col_name].round(3)
  return weights_df
  
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
linear_model_weights.sort_values(by=['features_df'])
```

RH_2 and RH_1 have the lowest and highest weights respectively

```python
# Ridge regression model

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

predicted_values = ridge_reg.predict(x_test)
```

```python
# Root Mean Squared Error after Ridge regression

from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3) 
```

No change in RMSE after train a ridge regression model

```python
# Lasso regression model

from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

predicted_values = lasso_reg.predict(x_test)
```

```python
# This function returns the features weights of the lasso model

def get_weights_df(lasso_reg, feat, col_name):
  weights = pd.Series(lasso_reg.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['features_df', col_name]
  weights_df[col_name].round(3)
  return weights_df

lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
lasso_weights_df
```

4 features have non-zero features weights

```python
# Root Mean Squared Error after Lasso regression

from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3) 
```

### Theory-based questions

The percent of the total variation of the dependent variable Y explained by the set of independent variables X is measured by Coefficient of Determination

Heteroskedasticity is linear regression with varying error terms

Linear regression is generally used for predicting continuous dependent variables

Lasso regression uses subset selection of features

Linear regression is sensitive to outliers

R Squared,Adjusted R Squared,F Statistics, RMSE / MSE / MAE metrics can be used for evaluating regression models

A best fit line relating X and Y has a R-Squared value of 0.75 means that 75% of the variance in Y is explained by X

R-square is optimal for comparing the goodness of the fit of competing regression models involving the same dependent variable

The Lasso can be interpreted as least-squares linear regression where weights are regularized with the L1 norm
